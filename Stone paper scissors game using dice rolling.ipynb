```python
import random
```

```python
def p_choice(p_number):
    valid_choices = ['rock', 'paper', 'scissors']
    while True:
        choice = input(f"Player {p_number}, choose rock, paper, or scissors: ").lower()
        if choice in valid_choices:
            return choice
        else:
            print("Invalid choice. Please try again.")
```

```python
def p_round(player1, player2):
    if player1 == player2:
        return 0 
    elif (player1 == 'rock' and player2 == 'scissors') or (player1 == 'paper' and player2 == 'rock') or (player1 == 'scissors' and player2 == 'paper'):
        return 1  
    else:
        return 2  
```

```python
def roll_dice():
    while True:
        dice_roll1 = random.randint(1, 6)
        dice_roll2 = random.randint(1, 6)
        print(f"Dice roll (player 1): {dice_roll1}")
        print(f"Dice roll (player 2): {dice_roll2}")
        if dice_roll1 == 6:
            print("Player 1 starts first!")
            return 1
        elif dice_roll2 == 6:
            print("Player 2 starts first!")
            return 2
#         else:
#             print("Neither player rolled a 6. Roll the dice again!")
#             roll_dice()
#             return
```

```python
print("----Welcome to the Stone-Paper-Scissors game!----")
print("----Both players will roll a dice, and the player who rolls a 6 will start first.----")

player1_score = 0
player2_score = 0

for r in range(1,11):
    starting_player = roll_dice()
    
    print(f"--- Round {r} ---")

    if starting_player == 1:
        player1_choice = p_choice(1)
        player2_choice = p_choice(2)
    else:
        player2_choice = p_choice(2)
        player1_choice = p_choice(1)

    result = p_round(player1_choice, player2_choice)

    if result == 1:
        print("Player 1 wins the round'..")
        player1_score += 1
    
    elif result == 2:
        print("Player 2 wins the round..")
        player2_score += 1
       
    else:
        print("It's a tie..")
        
        
print("--- Game Over ---")
print(f"Player 1 score: {player1_score}")
print(f"Player 2 score: {player2_score}")

if player1_score > player2_score:
    print("Player 1 wins the game...!")
elif player2_score > player1_score:
    print("Player 2 wins the game...!")
else:
    print("The game ends in a tie...!")
```
