```python
contacts={}
```

```python
def Create():
    name=input('Enter contact name: ').lower()
    if name not in contacts:
        number=input('Enter contact number: ')
        if number not in contacts:
            contacts.update({name:number})
            print('Contact added successfully') 
        else:
            print('Contact already exists')
    else:
        print('This name has already present in contacts. please enter different name')
        Create()
```

```python
def Search():
    name=input("Enter contact name to search: ").lower()
    if name in contacts:
        print("Contact found..")
        print("Name: ", name)
        print("Number: ", contacts[name])
    else:   
        print('Contact not found')
```

```python
def Update_number():
    name = input("Enter contact name to update: ")
    if name in contacts:
        number = input("Enter new contact number: ")
        contacts.update({name:number})
        print("Contact updated successfully")
    else:
        print("Contact not found")
```

```python
def Update_name():
    name = input("Enter name: ")
    if name in contacts:
        new_name = input("Enter new name: ")
        num = contacts[name]
        del contacts[name]
        contacts.update({new_name:num})
        print("Contact name updated successfully")
    else:
        print("Contact not found")
```

```python
def Delete():
    name=input('Enter contact name to delete: ').lower()
    if name in contacts:
        del contacts[name]
        print('Contact deleted successfully')
    else:
         print("Contact not found")
```

```python
def Display():
    if len(contacts)>0:
        print("Contact List:")
        for name, number in contacts.items():
            print("Name: ", name)
            print("Number: ", number)
            print("*******************")
    else:
        print("Contact list is empty")
```

```python
while True:
    print('****_____CONTACT BOOK MANAGEMENT_____****')
    print('1.CREATE NEW CONTACT')
    print('2.SEARCH THE CONTACT')
    print('3.UPDATE CONTACT NUMBER')
    print('4.DELETE THE CONTACT')
    print('5.UPDATE CONTACT NAME')
    print('6.DISPLAY THE CONTACT')
    print('7.Exit')
    Ch=input("Enter choice 1 to 7:")
    if Ch=='1':
        print("Create new contact...")
        Create()
    elif Ch=='2':
        print("Search contact...")
        Search()
    elif Ch=='3':
        print("Update contact number..")
        Update_number()
    elif Ch=='4':
        print("Delete contact...")
        Delete()
    elif Ch=='5':
        print("Update contact name...")
        Update_name()
    elif Ch=='6':
        print("Display contact...")
        Display()
    elif Ch=='7':
        print("Exiting...")
        break
    else:
        print('Invalid choice.please try again')
        
print('Thank you for visiting here...')
```
