```python
def Audience_Poll(i,m):
    L=[]
    import random as rd
    for x in range(20):
        L.append(rd.randint(1,4))
    #print(L)
    aver_score=sum(L)/20
    aud_ans=round(aver_score)
    print('Audiance choose option',aud_ans)
    go_with=input('Are you going with audience or quit:').lower()
    if go_with=='yes':
        if aud_ans==ans[i]:
            print(f'congratulations you give right answer and you win {pz[i]} rupees')
            m=pz[i]
            return m
        else:
            print(f'you give wrong answer, correct answer is {ans[i]}')
            return 'Au'
    else:
        return 'q'
        
#AudiencePoll()
```

```python
def validate_answer(i,answer,m3):
    if answer==ans[i]:
        print(f'congratulations you give right answer and you win {pz[i]} rupees')
        m3=pz[i]
        return m3
    else:
        print(f'you give wrong answer, correct answer is {ans[i]}')
        return 'W'
```

```python
def all_lifeline(i,money,answer):
    if lifeline == '50-50':
        if lifelines[lifeline]==1:
            lifelines[lifeline]=0
            M=fifty(i,money)
            if M=='l':return 0
            else:     return M
        else:
            print("Sorry, you have already used this lifeline. Please answer the question.")
            answer = int(input("Your answer (1/2/3/4): "))
            v1=validate_answer(i,answer,money)
            if v1=='W':return 0
            else:      return v1
                
    elif lifeline=='flip the question':
        if lifelines[lifeline]==1:
            lifelines[lifeline]=0
            Fl=Flip_Question(i,money)
            if Fl=='F':return 0
            else:      return Fl
        else:
            print("Sorry, you have already used this lifeline. Please answer the question.")
            answer = int(input("Your answer (1/2/3/4): "))
            v2=validate_answer(i,answer,money)
            if v2=='W':return 0
            else:      return v2
                
    elif lifeline=='audience poll':
        if lifelines[lifeline]==1:
            lifelines[lifeline]=0
            Aud=Audience_Poll(i,money)
            if Aud=='Au' or Aud=='q':return 0
            else:                    return Aud
        else:
            print("Sorry, you have already used this lifeline. Please answer the question.")
            answer = int(input("Your answer (1/2/3/4): "))
            v3=validate_answer(i,answer,money)
            if v3=='W':return 0
            else:      return v3
```
